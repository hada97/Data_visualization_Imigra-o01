#Visualizando Dados com a biblioteca matsplot

```python
import pandas as pd
```

Lendo arquivo csv e imprimindo os 5 primeiros com head()

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
df.head()
```

A

```python
df.info()
```

Definindo a coluna "país" como index

```python
df.set_index('País', inplace=True)
```

Cria uma lista com o intervalo de anos

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dic = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dic)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

Imprimindo gráfico e formatando a exibição nos eixos x e y

```python
plt.figure(figsize=(7, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigrantes do Brasil para Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

Usando figuras, as funções de formatação se tornam mais simples

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para Canadá de 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.yaxis.set_major_locator(plt.MultipleLocator(500))
plt.show()
```

Configurando 2 gráficos, o segundo usando boxplot

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá de 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para Canadá')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()


```

Analisando com 4 gráficos

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n 1980 a 2013', fontsize=16)

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')


for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))


for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')


ymin =0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

Agora vamos comparar todos os paises da América do Sul

```python
df.head()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(america_sul.index, america_sul['Total'])
```

Vamos adicionar cores ao grafico

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul.index, america_sul['Total'])
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=14)
plt.show()
```

Ordenando o DataFrame para ter um gráfico ordenado

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'])
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=14)
plt.show()
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=14)
plt.show()
```

Adicionando anotações e limpando a saída

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('blue')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá', loc = 'left', fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', ha ='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', length=0)

fig.savefig('migracao_america_sul.png', transparent=False, dpi=350, bbox_inches='tight')

plt.show()
```
